This nb demonstrates the use of a scipy stats function to plot probability plots. 
**scipy.stats.probplot(x, sparams=(), dist='norm', fit=True, plot=None, rvalue=False)**
Note that the dataset can be uploaded from a local source.
Ref. General syntax: 
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.probplot.html
Names of possible pmf or pdf:
https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html

```python
data = pd.read_csv('brain_size.csv', sep=';', na_values=".")
data  
```

```python
X=data['Weight']
Y=data['Height']
```

```python
# probability plot for one variable
# stats.probplot use the default distribution normal prob if not specified
fig = plt.figure()
ax = fig.add_subplot(111)
res = stats.probplot(X, plot=ax)
ax.set_title("Probplot for Variable X")
plt.show()
```

```python
# probability plot for one variable
# stats.probplot use the default distribution normal prob if not specified
fig = plt.figure()
ax = fig.add_subplot(111)
res = stats.probplot(X, plot=plt) #here the plot is plt; there is no difference in plots
ax.set_title("Probplot for Variable X")
plt.show()
```

```python
# probability plot for one variable
# stats.probplot tests loggamma distrubution 
fig = plt.figure()
ax = fig.add_subplot(111)
res = stats.probplot(Y, dist=stats.loggamma, sparams=(2.5,),plot=ax)
ax.set_title("Probplot for Variable Y")
plt.show()
```

```python
ax1 = plt.subplot(221)
res = stats.probplot(X, plot=plt)
ax2 = plt.subplot(222) #additional axis forces two plots
res = stats.probplot(Y, plot=plt)
plt.show()
```
